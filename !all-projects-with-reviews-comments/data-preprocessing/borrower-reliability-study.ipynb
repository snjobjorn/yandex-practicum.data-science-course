### Лемматизация

Определим уникальные значения столбца purpose.

```python
purpose_unique = pd.unique(data['purpose'])
print(purpose_unique)
```

Лемматизация уникальных значений столбца purpose.

```python
from pymystem3 import Mystem
m = Mystem()

lemmas = []
for i in purpose_unique:
    lemma = m.lemmatize(i)
    lemmas.append(lemma)

lemmas_united = [j for i in lemmas for j in i]

from collections import Counter
Counter(lemmas_united)
```

**Вывод**

Исходя из лемматизированных данных мы можем четко определить категории целей получения кредита.

### Категоризация данных

Принцип категоризации данных: Сужение спектра категорий в столбце purpose за счет объединения похожих, исходя из лемматизации и стемминга.

```python
def purpose_change(purpose):
    lemmas_row = m.lemmatize(purpose)
    for i in lemmas_row:
        if 'авто' in i:
            return 'автомобиль'
        if 'недвиж' in i:
            return  'недвижимость'
        if 'свад' in i:
            return  'свадьба'
        if 'образов' in i:
            return 'образование'
        if 'жиль' in i:
            return 'недвижимость'

data['purpose_grouped']= data['purpose'].apply(lambda x: purpose_change(x))     
data['purpose_grouped'].value_counts()
data.head(10)
```
